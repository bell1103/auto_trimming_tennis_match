# Display video 
Before running the trimming pipline we we'll visualize the input video

```python
from IPython.display import Video, display
import os

if not os.path.exists(VIDEO_PATH):
    print(f"Video at found at:  {VIDEO_PATH}")
else: 
    print(f"Previewing input video: {VIDEO_PATH}")
    display(Video(VIDEO_PATH, embed=True, width=800))
```

```python
trail_flag = "--display_trail" if DISPLAY_TRAIL else ""

!python main.py \
    --video_path "{VIDEO_PATH}" \
    --output_dir "{OUTPUT_DIR}" \
    --model_path "{MODEL_PATH}" \
    {trail_flag}
```
